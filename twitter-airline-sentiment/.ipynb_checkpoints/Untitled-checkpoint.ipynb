```python
switchindex = 0
switch = []
for text in data["text"]:
    text = re.sub(r'@', '', text)
    text = re.sub(r'[Uu][Nn][Ii][Tt][Ee][Dd]', '', text)
    text = re.sub(r'[Vv][Ii][Rr][Gg][Ii][Nn]\s*[Aa][Mm][Ee][Rr][Ii][Cc][Aa]', '', text)
    text = re.sub(r'@VirginAmerica', '', text)
    text = re.sub(r'[Ss][Oo][Uu][Tt][Hh][Ww][Ee][Ss][Tt]', '', text)
    text = re.sub(r'[Dd][Ee][Ll][Tt][Aa]', '', text)
    text = re.sub(r'[Uu][Ss]\s*[Aa][Ii][Rr][Ww][Aa][Yy][Ss]', '', text)
    text = re.sub(r'[Aa][Mm][Ee][Rr][Ii][Cc][Aa][Nn]', '', text)
    switch.append(text)

```

```python
new_series = pd.Series(switch)
new_series
```

```python
data["text"] = new_series
```

```python
for i in range(0, 14640):
    if data.loc[i, "airline"] == "Virgin America":
        if data.loc[i, "airline_sentiment"] == 'positive':
            Virgin_America_text_pos.append(data.loc[i, "text"])
            Virgin_America_labels_pos.append('positive')
        elif data.loc[i, "airline_sentiment"] == 'negative':
            Virgin_America_text_neg.append(data.loc[i, "text"])
            Virgin_America_labels_neg.append('negative')
    if data.loc[i, "airline"] == "United":
        if data.loc[i, "airline_sentiment"] == 'positive':
            United_text_pos.append(data.loc[i, "text"])
            United_labels_pos.append('positive')
        elif data.loc[i, "airline_sentiment"] == 'negative':
            United_text_neg.append(data.loc[i, "text"])
            United_labels_neg.append('negative')
    if data.loc[i, "airline"] == "Southwest":
        if data.loc[i, "airline_sentiment"] == 'positive':
            Southwest_text_pos.append(data.loc[i, "text"])
            Southwest_labels_pos.append('positive')
        elif data.loc[i, "airline_sentiment"] == 'negative':
            Southwest_text_neg.append(data.loc[i, "text"])
            Southwest_labels_neg.append('negative')  
    if data.loc[i, "airline"] == "Delta":
        if data.loc[i, "airline_sentiment"] == 'positive':
            Delta_text_pos.append(data.loc[i, "text"])
            Delta_labels_pos.append('positive')
        elif data.loc[i, "airline_sentiment"] == 'negative':
            Delta_text_neg.append(data.loc[i, "text"])
            Delta_labels_neg.append('negative')
    if data.loc[i, "airline"] == "US Airways":
        if data.loc[i, "airline_sentiment"] == 'positive':
            usairway_text_pos.append(data.loc[i, "text"])
            usairway_labels_pos.append('positive')
        elif data.loc[i, "airline_sentiment"] == 'negative':
            usairway_text_neg.append(data.loc[i, "text"])
            usairway_labels_neg.append('negative')
    if data.loc[i, "airline"] == "American":
        if data.loc[i, "airline_sentiment"] == 'positive':
            american_text_pos.append(data.loc[i, "text"])
            american_labels_pos.append('positive')
        elif data.loc[i, "airline_sentiment"] == 'negative':
            american_text_neg.append(data.loc[i, "text"])
            american_labels_neg.append('negative')
```

```python
training_text = []
training_labels = []

test_text = []
test_labels = []

positive_texts = [Virgin_America_text_pos, United_text_pos, Southwest_text_pos, Delta_text_pos, usairway_text_pos, american_text_pos]
negative_texts = [Virgin_America_text_neg, United_text_neg, Southwest_text_neg, Delta_text_neg, usairway_text_neg, american_text_neg]

positive_labels = [Virgin_America_labels_pos, United_labels_pos, Southwest_labels_pos, Delta_labels_pos, usairway_labels_pos, american_labels_pos]
negative_labels = [Virgin_America_labels_neg, United_labels_neg, Southwest_labels_neg, Delta_labels_neg, usairway_labels_neg, american_labels_neg]

# for i,j,k,l in positive_texts, negative_texts, positive_labels, negative_labels:
#     training_text = i[:int((.8)*len(i))] + j[:int((.8)*len(j))]
#     training_labels = k[:int((.8)*len(k))] + l[:int((.8)*len(l))]
    
#     test_text = i[int((.8)*len(i)):] + j[int((.8)*len(j)):]
#     test_labels = k[int((.8)*len(k)):] + l[int((.8)*len(l)):]
    
for i in positive_texts:
    training_text += i[:int((.8)*len(i))] 
    test_text += i[int((.8)*len(i)):] 
for j in negative_texts:
    training_text += j[:int((.8)*len(j))]
    test_text += j[int((.8)*len(j)):]
for k in positive_labels:
    training_labels += k[:int((.8)*len(k))]
    test_labels += k[int((.8)*len(k)):]
for l in negative_labels:
    training_labels += l[:int((.8)*len(l))]
    test_labels += l[int((.8)*len(l)):]
    


    

    
```

```python
# shuffle data set
```

Making a vector object called "vectorizer" that parses through the text by each word, and converts the most frequent 1000 words into numbers. The numbers represent the number of times each tweet contains the word (one word per column).

```python
vectorizer = CountVectorizer(
    analyzer = 'word',
    lowercase = False,
    max_features = 1000
)
```

Next, we use the fit_transform() method to transform our corpus data into feature vectors. Since the input needed is a list of strings, we concatenate all of our training and test data.

```python
features = vectorizer.fit_transform(
    training_text + test_text)

features_nd = features.toarray() # for easy use
```

As a final step, we'll split the training data to get an evaluation set through scikit-learn's built-in cross validation method. All we need to do is provide the data and assign a training percentage (in this case, 80%). X is the frequency of words (vectorized), and y is the associated label.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test  = train_test_split(
        features_nd[0:len(training_text)], 
        training_labels,
        train_size=0.80, 
        random_state=1234)
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
```

```python
log_model = log_model.fit(X=X_train, y=y_train)
```

```python
y_pred = log_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```
